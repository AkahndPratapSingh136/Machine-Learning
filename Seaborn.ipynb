#Visualizing statistical relationships
Statistical analysis is a process of understanding how variables in a dataset relate to each other and how those relationships depend on other variables. Visualization can be a core component of this process because, when data are visualized properly, the human visual system can see trends and patterns that indicate a relationship.

We will discuss three seaborn functions in this tutorial. The one we will use most is relplot(). This is a figure-level function for visualizing statistical relationships using two common approaches: scatter plots and line plots. relplot() combines a FacetGrid with one of two axes-level functions:

scatterplot() (with kind="scatter"; the default)

lineplot() (with kind="line")

As we will see, these functions can be quite illuminating because they use simple and easily-understood representations of data that can nevertheless represent complex dataset structures. They can do so because they plot two-dimensional graphics that can be enhanced by mapping up to three additional variables using the semantics of hue, size, and style.



#Relating variables with scatter plots
The scatter plot is a mainstay of statistical visualization. It depicts the joint distribution of two variables using a cloud of points, where each point represents an observation in the dataset. This depiction allows the eye to infer a substantial amount of information about whether there is any meaningful relationship between them.

There are several ways to draw a scatter plot in seaborn. The most basic, which should be used when both variables are numeric, is the scatterplot() function.
```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
```

```python
data=sns.load_dataset('tips')
```

```python
data
```

```python
sns.relplot(x='total_bill',y='tip',data=data)
```

**The hue parameter**

In seaborn, the hue parameter determines which column in the data frame should be used for colour encoding. Using the official document for lmplot provided an example for this. Adding `hue="smoker" tells seaborn you want to colour the data points for smoker and non-smoker differently.

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',data=data)
```

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='sex',data=data)
```

**The size parameter**

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='sex',size='size',data=data)
```

**Using scatterplot**

```python
sns.scatterplot(x='total_bill',y='tip',hue='smoker',style='sex',data=data)
```

It can also tell you if your data is symmetrical, how tightly your data is grouped, and how your data is skewed.

#Jointplot

Jointplot creates a multi-plane figure that shows both the bivariate (or joint) relationship between two variable along with the univariate (or marginal) distribution of each on separate axes.

We can use jointplots in following 4 manners:


1.   Scatter Plots
1.   Hexbin Plots
2.   Kde Plots
2.   Reg plot



```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', kind='hex',data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', kind='kde', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', kind='reg', data=tips)
```

#Distplots

This are used to do univariate analysis on a continous/numerical data.

Distplots combines 3 things in one graph:

1.Histogram

2.kernel Density Enstimation(PDF)

3.Rugplot

```python
titanic=sns.load_dataset('titanic')
titanic.head()
```

```python
titanic['age'].fillna(titanic['age'].mean(), inplace=True)
```

```python
sns.distplot(titanic['age'])
```

```python
sns.distplot(titanic['age'],kde=False)
```

```python
sns.distplot(titanic['age'],hist=False)
```

```python
sns.distplot(titanic['age'], hist=False, kde=False, rug=True)
```

A rug plot is a plot of data for a single quantitative variable, displayed as marks along an axis. It is used to visualise the distribution of the data. As such it is analogous to a histogram with zero-width bins, or a one-dimensional scatter plot.

```python
sns.distplot(titanic[titanic['survived']==1]['age'])
```

```python
sns.distplot(titanic[titanic['survived']==0]['age'])
```

```python
sns.distplot(titanic[titanic['survived']==1]['age'])
sns.distplot(titanic[titanic['survived']==0]['age'])
```

```python
import time
```

```python
x=range(10000000)
y=range(10000000,20000000)
start_time=time.time()
c=[(x,y) for x,y in zip(x,y)]
print(time.time()-start_time)
```

```python
x=np.arange(10000000)
y=np.arange(10000000,20000000)
start_time=time.time()
c=[(x,y) for x,y in zip(x,y)]
print(time.time()-start_time) 
```
